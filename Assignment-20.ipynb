```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(text, output_filename):
    # Generate the WordCloud with increased output size and customization
    wordcloud = WordCloud(
        width=1200,               # Increased width
        height=600,               # Increased height
        background_color='white', # White background
        max_words=200,            # Increase the number of words displayed
        colormap='viridis',       # Use a colormap for better visualization
        contour_width=1,          # Add contour to the word cloud
        contour_color='black'     # Contour color set to black
    ).generate(text)
    
    # Display the WordCloud
    plt.figure(figsize=(15, 7))  # Increase figure size
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    
    # Save the WordCloud as an image
    wordcloud.to_file(output_filename)
    print(f"WordCloud image saved as {output_filename}")

# Test the function with the given text
text = 'data science machine learning artificial intelligence'
output_filename = 'wordcloud_large.png'

generate_wordcloud(text, output_filename)
```
